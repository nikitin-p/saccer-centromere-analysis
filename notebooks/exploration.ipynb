# Centromere Analysis Exploration

Interactive exploration using the centromere package.

```python
import pandas as pd
from centromere.io import load_config, load_data
from centromere.preprocess import remove_gaps, find_split_hits, merge_split_hits
from centromere.filter import by_length, by_gc
from centromere.align import align_all_parallel
from centromere.mutations import compute_consensus, compute_mutation_patterns
from centromere.viz import plot_length_violin, plot_gc_violin
```

```python
# Load config
config = load_config('../config.yaml')
config
```

```python
# Load and preprocess data
df = load_data(config['input']['path'], config['input']['format'])
df = remove_gaps(df)
df.head()
```

```python
# Filter
df = by_length(df, config['filter']['min_length'])
df = by_gc(df, config['filter']['min_gc'])
print(f'Filtered: {len(df)} rows')
```
